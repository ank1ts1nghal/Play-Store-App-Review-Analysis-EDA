Before delving into the provided dataset, allow me to provide a brief explanation of Exploratory Data Analysis (EDA).

**Exploratory Data Analysis (EDA) :-**

Exploratory Data Analysis (EDA) is a crucial step in the data analysis process. It involves examining and visualizing the dataset to gain insights, discover patterns, and identify relationships between variables. Through EDA, we can uncover key characteristics of the data, such as its distribution, central tendencies, and variations. This exploration helps us understand the data's structure, identify any data quality issues or outliers, and guide us in selecting appropriate statistical techniques and models for further analysis. By employing various statistical and visualization techniques, EDA aids in formulating hypotheses, validating assumptions, and generating actionable insights that drive decision-making in data-driven projects. In summary, EDA serves as the foundation for a comprehensive understanding of the data, enabling us to extract meaningful information and drive meaningful conclusions.

Here are the general steps involved in Exploratory Data Analysis (EDA):

1. Data Collection: Gather the relevant dataset for analysis, which may come from various sources such as databases, files, or APIs.

2. Data Cleaning: Preprocess the data to handle missing values, outliers, and inconsistencies. This step involves techniques like data imputation, outlier detection, and handling categorical variables.

3. Data Exploration: Perform initial exploration by examining the dataset's structure, dimensions, and summary statistics. This includes checking the data types, understanding variable distributions, and identifying potential issues.

4. Univariate Analysis: Analyze individual variables to gain insights into their distributions, central tendencies, and variations. This involves plotting histograms, box plots, and computing summary statistics.

5. Bivariate Analysis: Explore relationships between pairs of variables to identify correlations, dependencies, or patterns. This includes scatter plots, correlation matrices, and grouped analysis.

6. Multivariate Analysis: Investigate interactions between multiple variables to uncover complex relationships or dependencies. Techniques like heatmaps, cluster analysis, or dimensionality reduction can be applied.

7. Feature Engineering: Create new features or transform existing ones based on domain knowledge and insights gained from the data analysis. This step aims to enhance predictive power or simplify the modeling process.

8. Visualization: Utilize various visualizations such as bar charts, line plots, scatter plots, or heatmaps to effectively communicate patterns and relationships in the data.

9. Statistical Analysis: Apply statistical techniques, such as hypothesis testing or regression analysis, to derive meaningful conclusions and validate assumptions.

10. Iterative Process: EDA is often an iterative process where insights from initial analyses lead to further exploration and refinement of hypotheses or models.

11. Documentation and Reporting: Summarize the findings, insights, and key takeaways from the EDA process. This documentation helps stakeholders understand the data and make informed decisions.

We have to keep in mind that the specific steps and techniques employed in EDA may vary depending on the nature of the data and the objectives of the analysis.

The choice between using the mean or the median depends on the distribution of the ratings.

Mean: The mean is the average value of the ratings and is sensitive to extreme values (outliers). If the ratings are normally distributed and there are no significant outliers, using the mean can provide a representative estimate of the missing values. It can be a suitable choice to maintain the overall average rating.

Median: The median represents the middle value of the ratings when they are sorted in ascending or descending order. Unlike the mean, the median is not affected by extreme values or outliers. If the ratings are skewed or have a few extreme values, using the median can provide a more robust estimate of the missing values. It can be a suitable choice to maintain the central tendency of the ratings.


We will Create a histogram of the ratings column to visualize the distribution of ratings. This will give you us an overview of the ratings distribution and help us understand if it is approximately normal or skewed.

We will also create a Box plot to identify any potential outliers. The box plot will show the median, quartiles, and any outliers in the data.

#### Chart - 1 : Ratio of number of Paid apps and Free apps
